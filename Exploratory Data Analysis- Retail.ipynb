# Bivariate Analysis of Continuous-Categorical variables

```python
#pairplot on the basis of Ship Mode
sns.pairplot(df,hue='Ship Mode')
```

```python
#creating a piechart for ship mode wise sales, profit and discount

df_shipmode = df.groupby(['Ship Mode'])[['Sales', 'Discount', 'Profit']].mean()
```

```python
#pie chart on basis of ship mode
df_shipmode.plot.pie(subplots=True,
                     figsize=(18, 20), 
                     autopct='%1.1f%%')
```

```python
#pairplot based on category
sns.pairplot(df, hue='Category')
```

```python
#creating a piechart for category wise sales, profit and discount

df_category = df.groupby(['Category'])[['Sales', 'Discount', 'Profit']].mean()
```

```python
df_category.plot.pie(subplots=True,
                     figsize=(18, 20), 
                     autopct='%1.1f%%')
```

#####  Technology has higher sales and profit. Highest discount is on furniture.

```python
#creating pairplot based on sub category
sns.pairplot(df, hue='Sub-Category')
```

```python
# subcategory wise sales, profit and discount analysis
df_sub_category = df.groupby(['Sub-Category'])[['Sales', 'Discount', 'Profit']].mean()
df_sub_category.head(10)
```

```python
#piechart for subcategory wise sales analysis
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Sales'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Sales Analysis', fontsize = 30)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

##### Copiers have the highest sales, followed by machines. 

```python
#piechart for subcategory wise discount analysis
plt.figure(figsize = (15,15))
plt.pie(df_sub_category['Discount'], labels = df_sub_category.index, autopct = '%1.1f%%')
plt.title('Sub-Category Wise Discount Analysis', fontsize = 30)
plt.legend()
plt.xticks(rotation = 90)
plt.show()
```

##### Highest discount is provided on binders and lowest on labels.

```python
df_sub_category.sort_values('Profit')[['Sales','Profit']]
```

##### Highest profit is earned on Copiers and the least on fasteners. The company suffers the highest loss on tables, followed by bookcases.

```python
#pairplot based on state
sns.pairplot(df,hue='State')
```

```python
#state wise sales, profit, discount
df_state= df.groupby(['State'])[['Sales', 'Discount', 'Profit']].mean()
df_state.head(10)
```

```python
#sorting states on profit basis
df_state1=df_state.sort_values('Profit')
df_state1
```

##### The company earns highest profit from Vermont and the highest losses are from the state of Ohio.

```python
#sorting states on sales basis
df_state2=df_state.sort_values('Sales')
df_state2
```

##### The highest sales made by the company is in Wyoming , followed by Vemont.

```python
#sorting states on discount basis
df_state3=df_state.sort_values('Discount')
df_state3
```

```python
# Pairplot Based on Region
sns.pairplot(df,hue='Region')
```

```python
#regionwise sales, profit and discount
df_region = df.groupby(['Region'])[['Sales', 'Discount', 'Profit']].mean()
df_region
```

```python
#piechart for region wise sales, profit and discount
df_region.plot.pie(subplots=True, 
                   figsize=(18, 20), 
                   autopct='%1.1f%%',
                   labels = df_region.index)
```

##### The highest sales are made to South region. The company provides highest discount in Central region and earns the highest profit from Western region.
